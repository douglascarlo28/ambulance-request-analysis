```python
import requests;
```

```python
import pandas as pd;
```

```python
# Requisitando dados da API

url = 'http://dados.recife.pe.gov.br/api/3/action/datastore_search?resource_id=5aaf98cb-0a6a-46e7-8cbd-3fedd6d4fa4d&limit=1000'  
r = requests.get(url)
```

```python
result = r.json()
result
```

```python
### Convertendo JSON to Dataframe

pdJson = pd.read_json(r.content)
dataframe = pd.json_normalize(pdJson['result']['records'])
dataframe
```

```python
# Encontrando a quantidae de chamados por cidade
QtdChamadosPorCidade = dataframe.groupby(['municipio']).size().sort_values(ascending=False)
QtdChamadosPorCidade
```

```python
# Identificar regiões que possuem maior incidência de um certo tipo de problema de saúde, 
# para criar pequenos centros especializados nestes problemas de saúde para estas regiões

# Utilizada esta função para juntar duas colunas de um Dataframe no formato de tabela.

pd.crosstab(dataframe['municipio'], dataframe['tipo'])
```
